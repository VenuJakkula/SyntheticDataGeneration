```python
import seaborn as sns
from matplotlib import pyplot as plt
cont_cols = ['dt', 'switch', 'pktcount', 'bytecount', 'dur',
       'dur_nsec', 'tot_dur', 'flows', 'packetins', 'pktperflow',
       'byteperflow', 'pktrate', 'Pairflow', 'port_no', 'tx_bytes',
       'rx_bytes', 'tx_kbps', 'rx_kbps', 'tot_kbps', 'label']
plt.figure(figsize=(20,10))
sns.heatmap(cleanedDf[cont_cols].corr(),annot=True).set_title("Synthetic Data")
```

```python
labelCompDf = pd.DataFrame()
labelCompDf["sythentic_labels_count"] = cleanedDf.label.value_counts()
labelCompDf["sythentic_labels%"] = cleanedDf.label.value_counts(normalize=True)
labels = '0', '1'
explode = (0, 0.1) 
fig1, ax1 = plt.subplots(nrows=1,ncols=2,figsize=(20,7))
ax1[0].pie(labelCompDf["sythentic_labels_count"].values, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1[0].set_title("Synthetic Data Labels")
```
